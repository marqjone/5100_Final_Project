1. Write a function that takes a string of this form, extracts the alphabetic code nnnnn and the extension and returns a string of the form nnnnn.extension. So, for example, given ‘35538_YLTRR_csv’ it returns ‘YLTRR.csv’.

```python
df = pd.read_csv("codes.txt", sep=" ", header=None, names=["Codes"])
print(df)
```

```python
# To keep things from being messy I am writing this function to take in each code every time it is called, rather than go through the entire list every time it is called.
# I will later use a for loop to iterate through the codes.txt values.
# This way the function itself doesn't need to be modified to accomadate another text file of values matching codes.txt data

def string_extractor(code):
    digit_code = code[0:5]
    alph_code = code[6:11]
    ext_code = code[12:]

    # Create the file name
    file_new_name = alph_code + "." + ext_code

    # Get current directory path
    current_dir = os.getcwd()
    current_dir = current_dir.replace('\\', '/')
    path = current_dir + "/expanded/" + file_new_name

    # Create new directory if it doesn't exist (ignore if it does), create file, write to file
    os.makedirs(os.path.dirname(path), exist_ok=True)
    with open(path, 'w') as f:
        f.write(digit_code)
        
    return file_new_name
```

```python
# Iterate through the df objects in the Codes column and pass the codes into the above function
for i in range(len(df)):
    new_code = df.loc[i, 'Codes']
    string_extractor(new_code)
```
