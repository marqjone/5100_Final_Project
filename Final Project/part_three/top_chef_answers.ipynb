```python
# Database functions

def run_sql(sql, engine):
    "send sql to database given by engine and return the result"
    with engine.connect() as conn:
        result = conn.execute(sqla.text(sql))
    return result.all()

def df_select(sql, engine):
    "send sql to database given by engine and return the result as a pandas dataframe"
    with engine.connect() as conn:
        result = pd.read_sql(sqla.text(sql), con=conn)
    print(f"Retrieved {result.shape[0]} records")
    return result
```

Compare the number of times a female was placed in the top 3 with the number of times a female won. Does this number seem unusual? That is, are women who place in the top 3 less likely than expected to ultimately win the competition? Why or why not?

Make a table with one row for each first place winner containing the winnerâ€™s name, the season they won, and a column containing â€˜Yesâ€™ or â€˜Noâ€™ depending on whether they served as a guest judge or not.

```python
new_table = pd.DataFrame()
mask = chef_det_interested['placement'].values == 1
df_new = chef_det_interested.loc[mask]
df_names = df_new["name"].values.tolist()
df_season = df_new["season"].values.tolist()

yes_no = []
for i in range(len(df_names)):
    true_false = (judges_interested['guestJudge'].eq(df_names[i])).any()
    if true_false == True:
        yes_no.append('Yes')
    else:
        yes_no.append('No')
    

new_table['WinnerName'] = df_names
new_table['Season'] = df_season
new_table['wasJudge'] = yes_no
display(new_table)
```

Find all contestants who were not first place winners but did serve as a guest judge at some point. Make sure to only list each name once.

```python
mask2 = chef_det_interested['placement'].values != 1
df_mask = chef_det_interested.loc[mask2]
non_winners = df_mask['name'].values.tolist()

judge_serve = []
for i in range(len(non_winners)):
    true_false = (judges_interested['guestJudge'].eq(non_winners[i])).any()
    judge_serve.append(true_false)

new_pd = pd. DataFrame({"name": non_winners,
    "onceJudge": judge_serve})

df_where_true = new_pd["name"].where(new_pd["onceJudge"] == True)
df_where_true = df_where_true.dropna()

df_where_true.drop_duplicates(keep='first', inplace = True)
display(df_where_true)
print(df_where_true.shape)

```
